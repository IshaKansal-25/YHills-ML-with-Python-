```python
# Subplot

#plot1 :-
x=np.array([0,1,2,3])
y=np.array([4,2,6,1])
plt.subplot(1,2,1)    # (row , column , plot no)
plt.plot(x,y)
plt.title("SALES")

#plot2 :-
x=np.array([0,1,2,3])
y=np.array([7,2,6,1])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")    # supertitle (title for entire figure)
plt.show()
```

```python
# Scatter Plot - scatter() plots one dot for each observation.
#                It needs two arrats of the same length(one for x-axis and one for y-axis)

x=np.array([5,7,2,8,9,4,1,8,6,2,7,8,6,4,2])
y=np.array([45,85,26,31,27,10,60,48,52,78,40,36,94,58,50])

plt.scatter(x,y)
plt.plot()
```

```python
x=np.array([5,7,2,8,9,4,1,8,6,2,7,8,6,4,2])
y=np.array([45,85,26,31,27,10,60,48,52,78,40,36,94,58,50])
plt.scatter(x,y,color='hotpink')

x=np.array([5,7,6,1,2,8,0,1,8,6,4,4,6])
y=np.array([89,41,30,45,71,65,98,71,53,42,64,10,58])
plt.scatter(x,y,color='green')
plt.plot()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
colors=['red','yellow','black','blue','green']
size=[10,200,54,40,35]
plt.scatter(x,y,c=colors,s=size)
plt.show()
```
